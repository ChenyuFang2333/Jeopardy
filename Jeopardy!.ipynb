```python
import pandas as pd
jeopardy = pd.read_csv('jeopardy.csv')
pd.set_option('display.max_colwidth', None)
```

## 1. Data Cleaning and Tidying

```python
#CLeaning the white spaces in the columns
jeopardy.columns = ['show_number', 'air_date', 'round', 'category', 'value', 'question', 'answer']
```

```python
jeopardy.info()
```

```python
#Creating a new column that converts the "Value" column to floats.
jeopardy['value_float'] = jeopardy.value.apply(lambda x: 0 if x =='None' else float(x.split('$')[-1].replace(',', '')))
jeopardy
```

```python
# Creating a new column specifying the year of each show
jeopardy['year'] = jeopardy.air_date.apply(lambda x: int(x.split('-')[0]))
jeopardy
```

## 2. A function that filters the dataset for questions that contains all of the words in a list of words.

```python
def filters(list_of_words, df):
    df_filtered = df[df.apply(lambda row: all(word.lower() in row.question.lower() for word in list_of_words), axis = 1)]
    return df_filtered
```

```python
contain_king_england = filters(['King','England'], jeopardy).reset_index()
contain_king_england
```

```python
# Looking at the average value of questions that contain the word "King"
filters(['King'],jeopardy).value_float.mean()
```

## 3. A function that returns the count of the unique answers to all of the questions in a dataset. 

```python
def unique_answer_count(df, list_of_words):
    filtered_df = filters(list_of_words, df)
    return filtered_df.answer.value_counts()
```

```python
# Find all of the unique answers to the questions containing word 'king'
unique_answer_count(jeopardy, ['king'])
```

## 4. Investigate the ways in which questions change over time.

```python
#How many questions from the 90s use the word "Computer" compared to questions from the 2000s?
jeopardy90s = jeopardy[(jeopardy.year >= 1990) & (jeopardy.year < 2000)]
jeopardy00s = jeopardy[(jeopardy.year >= 2000) & (jeopardy.year < 2010)]
len(filters(['computer'], jeopardy90s)) #98
len(filters(['computer'], jeopardy00s)) #268
```

## 5. Is there a connection between the round and the category? 

```python
#Is it more likely to find certain categories, like "Literature" in Single Jeopardy or Double Jeopardy?
#Modifying the filters() function to be able to filter not only questions but also other columns in the dataframe.
def new_filters(list_of_words, df, column):
    df_filtered = df[df.apply(lambda row: all(word.lower() in row[column].lower() for word in list_of_words), axis = 1)]
    return df_filtered

single_jeopardy = jeopardy[jeopardy['round']=='Jeopardy!']
double_jeopardy = jeopardy[jeopardy['round']=='Double Jeopardy!']
new_filters(['literature'], single_jeopardy, 'category') #423 rows
new_filters(['literature'], double_jeopardy, 'category') #1054 rows
#it is more likely to find category 'literature' in double jeopardies than in single jeopardies.
```

## 6. Quiz Time!

```python
# Build a system to grab random questions and take input from users as answers and check the answers.
def try_jeopardy():
    sample = jeopardy.sample()
    print(sample.question.item())
    input_answer = input('What is ')
    if input_answer.lower() == sample.answer.item().lower():
        return 'Congradulations! You earned {}!'.format(sample.value.item())
    else:
        return 'Sorry, you lost {}. The answer is {}. Good luck next time!'.format(sample.value.item(), sample.answer.item())
```
